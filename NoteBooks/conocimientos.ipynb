## Extraccion de datos

Para la extraccion de datos, existen 3 formas

- URL(Universal Resource Locator): Los programas pueden pedir la informacion directra al DIR de donde se guarda en el servidor. Problemas, costes altos al servidor, si cambia la localizacion del archivo la URL falla, riesgo de seguridad al dejar la URL o base de datos abiertas.
  
- API(Aplicattion Programming Interface): Es mas utilizado para conectar 2 tipos de programas como el front end con el back end, o distintos programs que no tengan nada que ver, solicitando datos o procesos entre si. Es alta mente utilizado.
  
- Web Scrapin(): Descarga el codigo de la pagina web con los dataos, mediante un escaneo de la pagina. 


Muchas veces no podemos elegir el proceso, sino nos debemos adaptar a las practicas del servidor donde se guardan los datos. En nuestro caso al quere evaluar datos del DEMRE, vimos que la opcion mas sencilla fue la descarga manual, la cual no la describi anteriormente porque es solamente hacer click en el enlace o boton que dice Descargar datos, no es nada tecnico ni muy avanzado. Luego los recopilamos en la misma carpeta y lo subimos a nuestro OneDrive para que puedan descargarlo, pero igualmente pueden ir y descargar los datos manualmetne, para tener certeza con lo que trabajamos.
