## ¿Cuántas personas se inscribieron en cada periodo de admisión entre 2004 y 2025?


### Librerias

    Utilizamos pickle para la apertura optima del data frame guardado y limpiado anteriormente. Matploit para la graficacion de los datos.

```python
import pickle
import matplotlib.pyplot as plt
```

### Load Dataframe
    Se carga los datos en la variable adm(admision).

```python
with open('pkl-data/adm_lista.pkl', 'rb') as lectura:
    adm = pickle.load(lectura)
```

### Codigo 

```python
tmp =[]
for x in range(len(adm)):
    tmp.append(int(adm[x].shape[0]))
print(tmp)
```

#### Explicación
    Al ser la primera pregunta, realizamos un codigo sencillo y eficiente, para dar el inicio a nuestro trabajo de investigacion. 
    
    Se recorre la lista, la cual tiene los data frame, y luego se crea la lista tmp(temporal), donde guardo la cantidad de filas, que alfinal son las entradas o cantidad de gente que rindo la prueba cada año. Al tener la lista, grafico los numeros de ella.


```python
#tmp = [153982, 169888, 176778, 212889, 218261,
#       243809, 253236, 252402, 242661, 244707,
#       243596, 257852,266494, 269351, 274498,
#       278287, 292091,238818, 254191, 293865,
#       306497, 311084]

years = list(range(2004, 2026))


plt.figure(figsize=(20, 10))
plt.plot(years, tmp, marker='o', linestyle='-', color='b')
plt.title('Valores por años 2004-2025', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Cantidad de gente que rindio la prueba', fontsize=12)
plt.xticks(years, rotation=45)
plt.grid(True)
plt.tight_layout()
plt.savefig('graficos-imagenes/pregunta1.png')# guardamos el grafico en su carpeta correspondiente

plt.show()
```

    - Aqui en el grafico, podemos ver la cantidad de gente que ha rendido la prueba de admision, a lo largo de los años.  
    - Logramos identificar claramente la cantidad de inscripcion por años, siendo el punto mas alto, el año actual 2025, con mas de 300.000 personas que rindieron el examen.

     
    Observaciones Importantes:
        
        -2010 to 2014, un claro descenso de la curva ascendente que teniamos los años pasadoso. Nuestra hipotesis mas logica es el terremoto del 2010 ocurrido en febrero, debio haber afectado claramente.
