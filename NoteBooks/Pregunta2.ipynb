##     ¿Hubo un aumento en las inscripciones en los periodos de transición (PSU/PTU/PAES)? 




### Librerias

    Utilizamos pickle para la apertura optima del data frame guardado y limpiado anteriormente. Matploit para la graficacion de los datos.

```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import os
```

### Load Dataframe
    Se carga los datos en la variable adm(admision).

```python
with open(os.path.join("pkl-data", 'adm_lista.pkl'), 'rb') as lectura:
    adm = pickle.load(lectura)
```

### Codigo 

    - El siguente codigo, fue inspirado y echo a base de la pregunta uno. Una extraccion exito de la cantidad de gente que rindio la prueba de admision superior a cada año. Con un cliclo for que recorre la lista de data frames, para luego la extraccion de la longitud que es equivalente a la cantidad de gente que rindo la prueba ese año. Luego fue graficado con matplpoit para la identificacion de maximos, minimos, pendientes, y posibles lugares que necesitaban mayor analisis. Fue subdividido para cada rango de años en los que se rindio cada tipop de prueba.


```python
# PSU del 2004 al 2020 como ultimo año
ruta = os.path.join("graficos-imagenes")
tmp =[]
for x in range(2004, 2021):
    x = x - 2004
    tmp.append(int(adm[x].shape[0]))
print(tmp)
years = list(range(2004, 2021))

plt.figure(figsize=(20, 10))
plt.plot(years, tmp, marker='o', linestyle='-', color='b')
plt.title('Valores por años PSU', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Gente Que Rindio La Prueba', fontsize=12)
plt.xticks(years, rotation=45)
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(ruta, 'pregunta2psu.png'))# guardamos el grafico en su carpeta correspondiente

plt.show()
```

```python
#PTU se hizo en el año 2021 y 2022
tmp =[238818, 254191]# extraccion manual de los datos pregunta 1, al ser solo dos datos era mas rapido,  facil y seguro.
years = [2021, 2022]

plt.figure(figsize=(20, 10))
plt.plot(years, tmp, marker='o', linestyle='-', color='b')
plt.title('Valores por años PTU', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Gente Que Rindio La Prueba', fontsize=12)
plt.xticks(years, rotation=45)
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(ruta, 'pregunta2ptu.png'))# guardamos el grafico en su carpeta correspondiente


plt.show()
```

```python
#PAES 2023 y 2025
tmp= [293865,306497, 311084]# extraccion manual de los datos pregunta 1, al ser solo dos datos era mas rapido,  facil y seguro.
years = [2023, 2024, 2025]

plt.figure(figsize=(20,10))
plt.plot(years, tmp, marker='o', linestyle='-', color='b')
plt.title('Valores por años PAES', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Gente Que Rindio La Prueba', fontsize=12)
plt.xticks(years, rotation=45)
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(ruta, 'pregunta2paes.png'))# guardamos el grafico en su carpeta correspondiente


plt.show()
```
